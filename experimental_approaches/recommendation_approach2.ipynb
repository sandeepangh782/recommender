```python
merged_data = pd.merge(interactions, meta, on='pratilipi_id', how='inner')
```

```python
merged_data
```

```python
merged_data['effective_read_time'] = merged_data['reading_time'] * (merged_data['read_percent'] / 100)
```

```python
merged_data
```

```python
user_last_interaction = merged_data.groupby('user_id')['updated_at'].max().to_dict()
```

```python
merged_data['days_since_user'] = merged_data.apply(
    lambda row: (user_last_interaction[row['user_id']] - row['updated_at']).days, axis=1
)
```

```python
merged_data
```

```python
import math

lambda_user = 0.1    # Decay rate for user-specific recency
lambda_publish = 0.1 # Decay rate for publication recency
alpha = 0.5
merged_data['r_ui_user'] = merged_data['days_since_user'].apply(lambda x: math.exp(-lambda_user * x)) # Use math.exp to calculate the exponential
```

```python
merged_data
```

```python
merged_data['days_since_publish'] = (merged_data['updated_at'] - merged_data['published_at']).dt.days
# Handle cases where interaction might (unexpectedly) occur before publication
merged_data['days_since_publish'] = merged_data['days_since_publish'].apply(lambda x: x if x >= 0 else 0)
# Compute publication recency factor (using exponential decay)
merged_data['r_ui_publish'] = merged_data['days_since_publish'].apply(lambda x: math.exp(-lambda_publish * x))
```

```python
merged_data
```

```python


merged_data.to_csv('merged_data.csv', index=False)
```

```python
merged_data['composite_recency'] = alpha * merged_data['r_ui_user'] + (1 - alpha) * merged_data['r_ui_publish']
```

```python
merged_data['composite_weight'] = merged_data['effective_read_time'] * merged_data['composite_recency']
```

```python
aggregated = merged_data.groupby(['user_id', 'pratilipi_id']).agg({
    'composite_weight': 'sum'
}).reset_index()
```

```python
merged_data_sorted = merged_data.sort_values('updated_at')
split_index = int(0.75 * len(merged_data_sorted))
train_data = merged_data_sorted.iloc[:split_index]
train_aggregated = train_data.groupby(['user_id', 'pratilipi_id']).agg({
    'composite_weight': 'sum'
}).reset_index()
```

```python
from lightfm import LightFM
from lightfm.data import Dataset
import numpy as np

dataset = Dataset()
dataset.fit(users=aggregated['user_id'].unique(), items=aggregated['pratilipi_id'].unique())
```

```python
train_interactions, _ = dataset.build_interactions(
    [(row['user_id'], row['pratilipi_id'], row['composite_weight']) for _, row in train_aggregated.iterrows()]
)
```

```python
meta['category_name'] = meta['category_name'].fillna('Unknown')
dataset.fit_partial(items=meta['pratilipi_id'].unique(), item_features=meta['category_name'].unique())
item_features = dataset.build_item_features(
    [(row['pratilipi_id'], [row['category_name']]) for _, row in meta.iterrows()]
)
```

```python
model = LightFM(loss='warp')
model.fit(train_interactions, item_features=item_features, epochs=30, num_threads=4)
```

```python
def recommend_pratilipis(model, dataset, user_id, item_features, num_recommendations=5):
    (user_map, item_map, _,_) = dataset.mapping()
    if user_id not in user_map:
        return []
    user_x = user_map[user_id]
    n_items = len(item_map)
    # Predict scores for all items for the given user
    scores = model.predict(user_x, list(range(n_items)), item_features=item_features)
    # Get the indices of the top scoring items
    top_items = sorted(range(n_items), key=lambda i: scores[i], reverse=True)[:num_recommendations]
    # Map back to the original pratilipi_ids
    inv_item_map = {v: k for k, v in item_map.items()}
    return [inv_item_map[i] for i in top_items]

# Example: Recommend pratilipis for a sample user
recommendations = recommend_pratilipis(model, dataset, user_id=5506791996685385, item_features=item_features)
print("Recommended pratilipis for user 1:", recommendations)
```
