```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from implicit.als import AlternatingLeastSquares
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
user_data = pd.read_csv('user_interaction.csv')
meta_data = pd.read_csv('metadata.csv')

print(f"User interaction data: {user_data.shape[0]} rows")
print(f"Meta data: {meta_data.shape[0]} rows")

# Convert timestamp columns to datetime
user_data['updated_at'] = pd.to_datetime(user_data['updated_at'])
meta_data['updated_at'] = pd.to_datetime(meta_data['updated_at'])
meta_data['published_at'] = pd.to_datetime(meta_data['published_at'])
```

```python
user_data['read_percent_norm'] = user_data['read_percent'] / 100.0

# 2. Get recency of interaction (days from latest date in dataset)
max_date = user_data['updated_at'].max()
user_data['days_since_interaction'] = (max_date - user_data['updated_at']).dt.days

# 3. Calculate recency weight (more recent = higher weight)
max_days = user_data['days_since_interaction'].max()
user_data['recency_weight'] = 1 - (user_data['days_since_interaction'] / max_days)

# 4. Combine read percentage and recency for a weighted engagement score
user_data['engagement_score'] = user_data['read_percent_norm'] * (0.7 + 0.3 * user_data['recency_weight'])

# 5. Merge with meta_data to get content features
enhanced_data = user_data.merge(meta_data, on='pratilipi_id', how='left')

# 6. Create reading time bins (short, medium, long stories)
reading_time_quantiles = meta_data['reading_time'].quantile([0.33, 0.66]).values
meta_data['length_category'] = pd.cut(
    meta_data['reading_time'], 
    bins=[0, reading_time_quantiles[0], reading_time_quantiles[1], float('inf')],
    labels=['short', 'medium', 'long']
)

# 7. Get content age in days
meta_data['content_age_days'] = (max_date - meta_data['published_at']).dt.days

# Print summary of engineered features
print("Added engagement score combining read percentage and recency")
print("Added content length categories")
print(f"Read percentage range: {user_data['read_percent_norm'].min():.2f}-{user_data['read_percent_norm'].max():.2f}")
print(f"Engagement score range: {user_data['engagement_score'].min():.2f}-{user_data['engagement_score'].max():.2f}")
```

```python
print("\n=== Preparing Model Data ===")

# Create a user-item interaction matrix
# We'll use the engagement_score as the interaction strength
user_ids = user_data['user_id'].unique()
item_ids = meta_data['pratilipi_id'].unique()

# Create mappings for user and item IDs
user_to_idx = {user: i for i, user in enumerate(user_ids)}
item_to_idx = {item: i for i, item in enumerate(item_ids)}
idx_to_user = {i: user for user, i in user_to_idx.items()}
idx_to_item = {i: item for item, i in item_to_idx.items()}

# Create the sparse matrix
rows = [user_to_idx[user] for user in user_data['user_id']]
cols = [item_to_idx.get(item, 0) for item in user_data['pratilipi_id']]
data = user_data['read_percent'].values / 100.0  # Normalize to 0-1

user_item_matrix = csr_matrix((data, (rows, cols)), shape=(len(user_ids), len(item_ids)))

print(f"Created user-item matrix with shape: {user_item_matrix.shape}")
print(f"Matrix density: {user_item_matrix.nnz / (user_item_matrix.shape[0] * user_item_matrix.shape[1]):.6f}")

# Temporal train-test split (75% train, 25% test)
# Sort by timestamp
user_data_sorted = user_data.sort_values('updated_at')

# Calculate the split index
split_idx = int(len(user_data_sorted) * 0.75)

# Split the data
train_data = user_data_sorted.iloc[:split_idx]
test_data = user_data_sorted.iloc[split_idx:]

print(f"Train data: {train_data.shape[0]} interactions")
print(f"Test data: {test_data.shape[0]} interactions")

# Create train matrix
train_rows = [user_to_idx[user] for user in train_data['user_id']]
train_cols = [item_to_idx.get(item, 0) for item in train_data['pratilipi_id']]
train_data_values = train_data['read_percent'].values / 100.0

train_matrix = csr_matrix((train_data_values, (train_rows, train_cols)), 
                          shape=(len(user_ids), len(item_ids)))

print(f"Train matrix shape: {train_matrix.shape}")

# Create mappings for category and author features
# These will be used for content-based recommendations
all_categories = meta_data['category_name'].unique()
category_to_idx = {cat: i for i, cat in enumerate(all_categories)}

# Create a pratilipi-category matrix
item_category_data = []
for _, row in meta_data.iterrows():
    item_idx = item_to_idx.get(row['pratilipi_id'])
    if item_idx is not None:
        category_idx = category_to_idx.get(row['category_name'])
        if category_idx is not None:
            item_category_data.append((item_idx, category_idx, 1.0))

item_cat_rows = [i for i, _, _ in item_category_data]
item_cat_cols = [j for _, j, _ in item_category_data]
item_cat_data = [v for _, _, v in item_category_data]

item_category_matrix = csr_matrix((item_cat_data, (item_cat_rows, item_cat_cols)), 
                                 shape=(len(item_ids), len(all_categories)))

print(f"Item-category matrix shape: {item_category_matrix.shape}")
```

```python
model_data={ 'user_item_matrix': user_item_matrix,
        'train_matrix': train_matrix,
        'test_data': test_data,
        'user_to_idx': user_to_idx,
        'item_to_idx': item_to_idx,
        'idx_to_user': idx_to_user,
        'idx_to_item': idx_to_item,
        'item_category_matrix': item_category_matrix,
        'category_to_idx': category_to_idx
    }
```

```python
print("\n=== Training Model ===")

# Initialize the ALS model
model = AlternatingLeastSquares(
    factors=100,        # Number of latent factors
    regularization=0.1, # Regularization parameter
    iterations=15,      # Number of iterations to run
    random_state=42
)

# Train the model
print("Training ALS model...")
model.fit(model_data['train_matrix'])
print("Model training complete")
```

```python
print("\n=== Generating Recommendations ===")

user_ids = list(model_data['idx_to_user'].keys())
item_ids = list(model_data['idx_to_item'].keys())

# Get recommendations for each user
recommendations = {}

for user_idx in tqdm(user_ids, desc="Generating recommendations"):
    scores = model.recommend(
        user_idx,
        model_data['train_matrix'][user_idx],
        N=5,  # num_recommendations
        filter_already_liked_items=True
    )
    
    # Unpack item indices and discard scores
    item_indices, _ = scores  
    user_id = model_data['idx_to_user'][user_idx]
    recommended_items = [model_data['idx_to_item'][item_idx] for item_idx in item_indices]
    
    recommendations[user_id] = recommended_items


print(f"Generated recommendations for {len(recommendations)} users")
```

```python
print("\n=== Evaluating Model ===")

test_data = model_data['test_data']
user_to_idx = model_data['user_to_idx']

# Create a dictionary of actual items for each user in the test set
test_user_items = {}
for user_id, group in test_data.groupby('user_id'):
    if user_id in user_to_idx:
        test_user_items[user_id] = set(group['pratilipi_id'])

# Calculate precision and recall
precisions = []
recalls = []

for user_id, actual_items in test_user_items.items():
    if user_id in recommendations:
        recommended_items = set(recommendations[user_id])
        
        # Items that are both recommended and in the test set
        true_positives = len(recommended_items.intersection(actual_items))
        
        # Precision: proportion of recommended items that are relevant
        precision = true_positives / len(recommended_items) if recommended_items else 0
        
        # Recall: proportion of relevant items that are recommended
        recall = true_positives / len(actual_items) if actual_items else 0
        
        precisions.append(precision)
        recalls.append(recall)

avg_precision = np.mean(precisions)
avg_recall = np.mean(recalls)

print(f"Average Precision: {avg_precision:.4f}")
print(f"Average Recall: {avg_recall:.4f}")
print(f"F1 Score: {2 * avg_precision * avg_recall / (avg_precision + avg_recall) if (avg_precision + avg_recall) > 0 else 0:.4f}")
```

```python
result={
        'precision': avg_precision,
        'recall': avg_recall,
        'user_coverage': len(precisions) / len(test_user_items)
    }
```

```python
print("\n=== Enhancing Recommendations with Content Features ===")

# For users with no collaborative filtering recommendations,
# we'll use content-based recommendations based on category preferences

# Get all users who don't have recommendations
all_users = set(model_data['user_to_idx'].keys())
users_with_recs = set(recommendations.keys())
cold_start_users = all_users - users_with_recs

print(f"Found {len(cold_start_users)} users without collaborative filtering recommendations")

# For these users, find their historical category preferences
enhanced_recommendations = recommendations.copy()

for user_id in cold_start_users:
    # Get the user's historical interactions
    user_idx = model_data['user_to_idx'][user_id]
    user_vector = model_data['train_matrix'][user_idx].toarray().flatten()
    
    # Get the items this user has interacted with
    user_items = [i for i, val in enumerate(user_vector) if val > 0]
    
    if not user_items:
        # If user has no items in training set, recommend popular items
        popular_items = meta_data['pratilipi_id'].value_counts().head(5).index.tolist()
        enhanced_recommendations[user_id] = popular_items
    else:
        # Get the categories of items the user has interacted with
        user_item_ids = [model_data['idx_to_item'][idx] for idx in user_items]
        user_categories = meta_data[meta_data['pratilipi_id'].isin(user_item_ids)]['category_name'].value_counts()
        
        # Get items in the user's favorite categories that they haven't seen
        if len(user_categories) > 0:
            top_category = user_categories.index[0]
            category_items = meta_data[meta_data['category_name'] == top_category]['pratilipi_id'].tolist()
            
            # Filter out items the user has already interacted with
            new_items = [item for item in category_items if item not in user_item_ids]
            
            # Take the top 5 or whatever is available
            recommended_items = new_items[:5] if len(new_items) >= 5 else new_items
            
            if len(recommended_items) < 5:
                # Fill remaining slots with popular items
                popular_items = meta_data['pratilipi_id'].value_counts().head(10).index.tolist()
                for item in popular_items:
                    if item not in recommended_items and item not in user_item_ids:
                        recommended_items.append(item)
                        if len(recommended_items) >= 5:
                            break
            
            enhanced_recommendations[user_id] = recommended_items[:5]
        else:
            # Fallback to popular items
            popular_items = meta_data['pratilipi_id'].value_counts().head(5).index.tolist()
            enhanced_recommendations[user_id] = popular_items

print(f"Enhanced recommendations for {len(cold_start_users)} cold-start users")
print(f"Total users with recommendations: {len(enhanced_recommendations)}")
```

```python
print("\n=== Formatting Final Recommendations ===")

# Create a dataframe with user_id, pratilipi_id pairs
output_rows = []

for user_id, items in recommendations.items():
    for item_id in items:
        output_rows.append({
            'user_id': user_id,
            'pratilipi_id': item_id
        })

output_df = pd.DataFrame(output_rows)

# Add category information
output_with_meta = output_df.merge(meta_data[['pratilipi_id', 'category_name']], on='pratilipi_id', how='left')

# Print sample of recommendations
print("\nSample recommendations:")
print(output_with_meta.head(10))

# Save to CSV
output_df.to_csv('pratilipi_recommendations.csv', index=False)
print(f"Saved {len(output_df)} recommendations to pratilipi_recommendations.csv")
```
